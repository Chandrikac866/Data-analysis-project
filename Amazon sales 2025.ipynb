```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('amazon_sales_data 2025.csv')
```

```python
data.drop(columns=['Order ID','Date'],inplace=True)
```

```python
data['Customer Name'].value_counts()
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=data,x='Customer Name',color='green')
plt.title("Count of the Customer ")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=data,x='Customer Name',hue='Category')

plt.show()
```

```python
sns.barplot(data=data,x='Payment Method',y='Total Sales',color='yellow')
plt.title("Total Sales by Price")
plt.show()
```

```python
sns.boxplot(data=data,x='Quantity',color='orange')
plt.title("Finding the outliers using box plot for the attribue Quantity")
plt.show()
```

```python
sns.barplot(data=data,y='Product',x='Status',hue="Category")
plt.title("Based on the Product check the Status by Categories of product")
plt.show()
```
