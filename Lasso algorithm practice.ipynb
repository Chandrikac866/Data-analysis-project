```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
data=pd.read_csv('loan_dataset_20000.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data.drop(columns=['loan_purpose','interest_rate','loan_term','installment','grade_subgrade','num_of_open_accounts','total_credit_limit','current_balance'
                  ,'delinquency_history','public_records','loan_paid_back','num_of_delinquencies','debt_to_income_ratio'],inplace=True)
```

```python
data
```

```python
x=data.iloc[:,:-1]
```

```python
x
```

```python
y=data['loan_amount']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
transformer=ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(sparse_output=False,drop='first'),['gender','marital_status','education_level','employment_status']
)],remainder='passthrough')
```

```python
model=Pipeline(steps=[
    ('transformer',transformer),
    ('Lasso',Lasso(alpha=1.0,fit_intercept=True))
])
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_train,y_train)*100,model.score(x_test,y_test)*100
```
