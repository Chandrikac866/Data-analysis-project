```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Birth rate of each country per year.csv')
```

```python
data.sample(10)
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data.rename(columns={"Births - Sex: all - Age: all - Variant: estimates":"birth rate"},inplace=True)
```

```python
#Entity vsbirth rate
#Year vs birth rate
#entity vs year 
```

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=(35,10))
sns.barplot(x='Year',y='birth rate',data=data)
plt.show()
```

```python
plt.figure(figsize=(10,40))
sns.barplot(y='Entity',x='birth rate',data=data,color='green')
plt.show()
```

```python
sns.lineplot(data=data,x='Year',y='birth rate')
plt.show()
```
